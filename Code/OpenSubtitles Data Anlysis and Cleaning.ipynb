```python
# Remove normal characters
def remnormals(sen):
    l = sen
    l = l.replace('.','').replace('?','').replace('!','').replace('\'','').replace(',','').replace('--','')
    l = l.replace('-','').replace('\n','').replace(';','').replace(':','').replace('\"','').replace('/','')
    l = l.replace('`','').replace("[",'').replace("]",'').replace("&",'')
    l = l.replace("#",'').replace("$",'').replace("+",'').replace("%",'').replace("^",'').replace('…','')
    l = l.replace("=",'').replace("´",'').replace("·",'').replace("~",'').replace("<",'').replace(">",'')
    l = l.replace("«",'').replace("»",'').replace("\\",'').replace("|",'').replace('\—','').replace("’",'')
    l = l.replace("“",'').replace("”",'').replace("\xad",'').replace("{",'').replace("}",'').replace("£",'')
    l = l.replace('‘','')

    return l
```

```python
spark = SparkSession.builder.appName("Sparkify").config('spark.sql.session.timeZone', 'GMT+1') \
        .getOrCreate()
```

```python
# Load in previous dictionairy of dataframes, that need to be converterd to spark dataframes
with open('newdfs.pkl', 'rb') as fp:
    newdfs = pickle.load(fp)
newdfs[list(newdfs.keys())[0]]
```

```python
%%time
keys = list(newdfs.keys())
# To Pandas is soly done for visualisation purposes
for k in keys:
  newdfs[k] = spark.createDataFrame(newdfs[k]).withColumn('Sentence', regexp_replace('Sentence','\n', ''))\
    .withColumn('Speaker', regexp_replace('Speaker',"[\n|\t]|(continued)", ''))\
        .withColumn('Next Speaker', regexp_replace('Next Speaker',"[\n|\t]|(continued)", ''))\
            .withColumn("Y", when(col("Speaker") == col("Next Speaker"),0).otherwise(1))\
                .withColumn("Sentence Length",F.split('Sentence',' ')).withColumn('Sentence Length',size('Sentence Length')).toPandas()

newdfs[keys[0]]
```
